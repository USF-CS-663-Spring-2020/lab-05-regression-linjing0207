```python
import pandas as pd

df = pd.read_csv('Credit.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

### Target: Limit, Predictors: Income

```python
import matplotlib.pyplot as plt

df.plot(kind="scatter", x="Income", y="Limit")
```

```python
import seaborn as sns

sns.lmplot(x='Income', y='Limit', data=df)
```

```python
r = sns.residplot('Income','Limit', data=df)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(df[['Income']], df['Limit'])
lin_reg.coef_, lin_reg.intercept_
```

```python
# Conclusion: The points in a residual plot are randomly dispersed around the horizontal axis, 
# so this linear regression model is appropriate for the data. The predictor Income is suitable for SLR.
# Scope is 51.88, intercept is 2389.87. 
```
